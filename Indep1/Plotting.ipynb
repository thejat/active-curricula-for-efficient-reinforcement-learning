```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import savefig
```

```python
data = np.load("../base2.npz")
data = data['STEP']
rounds = data.shape[0]
obs = data[0].shape[0]
# print obs
std = np.zeros(obs)
x = np.zeros(obs)
for i in range(obs):
    a = np.array([])
    for j in range(rounds):
        a = np.append(a, data[j][i])
    x[i] = np.mean(a)
    std[i] = np.std(a)
# print x, std
std = std/np.amax(x)
x  = x/np.amax(x)
y = np.ones(27)
y2 = np.arange(24)
y3 = np.arange(27)
y3 = y3-1
```

```python
# plt.rcParams.update({'font.size': 12})
plt.plot(x[:-1], linewidth = 1.5, color = "b", marker = 'o', label = "All curriculum")
# plt.plot(x[:-1], 'bo' )
plt.errorbar(y2, x[:-1], std[:-1], linestyle='None', color = 'black')
plt.plot(y3, y, color="red", linewidth = 2, label="baseline")
plt.plot([0,1], x[0:2], 'ys', label = "curriculum selected")#' color="red", linewidth = 2, label="baseline")
plt.legend(loc=4)
plt.axis([-1, 25, 0.4, 1.1])
plt.ylabel("Fraction of steps required for convergence", fontsize = 12)
plt.xlabel("Curriculum Number", fontsize = 12)
plt.show()
plt.savefig("res1.png")
```
